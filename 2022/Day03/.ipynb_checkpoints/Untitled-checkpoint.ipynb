```python
#function to open the input file and iterate trought it if there is a empty line it creates a new elve 
def open_file(file_name):
    elves = []
    elves_counter = 0
    calories_list = []
    calories_list.clear()
    with open(file_name, 'r') as f:
        
        for line in f:
            if len(line)>1:
                line = line.strip()
                line = int(line)
                calories_list.append(line)
            else:
                elves.append(Elve(name=elves_counter,calories=calories_list))
                elves_counter+=1
                calories_list.clear()
                
    return elves
                
#new elves class that stores a list with all calories in it
class Elve:
    def __init__(self, name, calories):
        self.name = name
        self.calories = calories
        print("New Instance of Elve with name: ", self.name)
    def __str__(self):
        return f'Elve {self.name} List: {self.calories} Total: {self.total_calories()}'
    
    def total_calories(self):
        return sum(self.calories)

        
def day01(file_name):
    elves = open_file(file_name)
    for elve in elves:
        print(elve)

        
day01("input.txt")
```
